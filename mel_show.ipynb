导入库

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
```

加载音频文件

```python
audio_path = '/private/tmp/芙宁娜/vo_FDAQ001_5_furina_17.wav'  # 替换为你的音频文件路径
y, sr = librosa.load(audio_path, sr=None)
print(f"音频采样率: {sr}\t 音频时长: {len(y) / sr:.2f} 秒")
```

设置参数

```python
n_fft = 2048
hop_length = 512
n_mels = 128
win_length = 2048
```

计算梅尔频谱图

```python
mel_spectrogram = librosa.feature.melspectrogram(y=y, sr=sr, n_fft=n_fft, hop_length=hop_length, n_mels=n_mels, win_length=win_length)
```

转换为分贝单位

```python
mel_spectrogram_db = librosa.power_to_db(mel_spectrogram, ref=np.max)
```

显示梅尔频谱图

```python
plt.figure(figsize=(10, 6))
librosa.display.specshow(mel_spectrogram_db, sr=sr, hop_length=hop_length, x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f dB')
plt.title('Mel Spectrogram')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.show()
```
